```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(-3,3,size=100)
```

```python
X = x.reshape(-1,1)
```

```python
y = 0.5*x**2 + x +2 +np.random.normal(0,1,size=100)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X,y)
```

```python
y_predict = lin_reg.predict(X)
```

```python
plt.scatter(x,y)
plt.plot(x,y_predict,color='r')
plt.show()
```

### 解决方案 添加一个特征

```python
(X**2).shape
```

```python
X2 = np.hstack([X,X**2])
```

```python
X2.shape
```

```python
lin_reg2 = LinearRegression()
```

```python
lin_reg2.fit(X2,y)
```

```python
y_predict2 = lin_reg2.predict(X2)
```

```python
plt.scatter(x,y)
plt.plot(x,y_predict2,color='r')
plt.show()
```

```python
plt.scatter(x,y)
plt.plot(np.sort(x),y_predict2[np.argsort(x)],color='r')
plt.show()
```

```python
lin_reg2.coef_  #[X,X**2] 第一个是x前的系数，第二个是x**2 前的系数
```

```python
lin_reg2.intercept_
```

### 升维思路，添加一个特征
